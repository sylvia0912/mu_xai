## XAI Interpretibility Techniques 

#### 6 Categories-
1. Intrinsic:  White-box models r inherently transparent - e.g. regression & decision tree.
2. Post-Hoc:  Applied after model training, applicable to both white & black box.
3. Model-Specific:  technique is tied to specific structure of model being used.
4. Model-Agnostic:  applied to any model, independent of internal structure.
5. Local:  Give explanation for specific instance.
6. Global:  Give explanation for model as a whole.
