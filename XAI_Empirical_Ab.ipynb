## Local Linear Model



1. Generate random datapoints in neighbourhood of input data point.
2. Get predictions on these data points using complex model (ensemble / neural network). Now u have a labelled new dataset created around the local data point which is to be predicted.
3. Then train simple linear model on this (labelled data by complex model).
4. Actual value/ground truth by complex model compared with predicted value obt. from simple model for training this linear model.
