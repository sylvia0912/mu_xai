### Imbalance in Classification and Regression Tasks

Feature importance scores can be calculated for problems that involve predicting a numerical value, called regression, and those problems that involve predicting a class label, called classification.

To study effect of Data Imbalance w.r.t Classification or Regression task ??

https://towardsdatascience.com/data-imbalance-in-regression-e5c98e20a807

- **Classification Task**<br>
Data imbalance is a typical problem for real world data sets. Data imbalance can be best described by looking at a binary classification task. In binary classification, the data set is imbalanced if the amount of samples between classes zero and one are uneven. The machine learning model tends to be better at predicting the class with more samples (majority class) than the one with fewer samples (minority class). **The greater this imbalance, the higher the bias of the model towards the majority class.**


- **Regression Task**<br>
In regression, the data imbalance is not quite as obvious, but can be present and cause problems. In the optimal case, the distribution of the target variable would be normally distributed. But, as one can see, this is clearly not the case for the target variable of the Boston Housing data set.The mean is around 20 and the first part already looks quite like a normal distribution. But there is a large right tail of higher MEDV values. This could lead to the problem, that the model better predicts the MEDV values around the mean, but is quite bad at predicting the MEDV values from the right tail. This is because most of the time the model sees values around the mean and is therefore biased towards these MEDV values.A statistical test can also be executed in order to verify that this distribution is not a normal distribution.
